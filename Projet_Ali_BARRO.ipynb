```python
#librairies 
import urllib.request
import pandas as pd
import numpy as np
import zipfile
import os
```

```python
# 1.Après lecture,le Code ISO pour le Burkina Faso est BF
iso_code_burkina_faso = 'BF'

# 2 - Télécharger le fichier zip correspondant(il faut être connecté à internet pour pouvoir télécharger les données)
url_burkina_faso = f'https://download.geonames.org/export/dump/{iso_code_burkina_faso}.zip'
urllib.request.urlretrieve(url_burkina_faso, 'BF.zip')

print("Téléchargement du fichier du Burkina Faso terminé.")
```

```python
#3.
# Chemin vers le fichier ZIP
zip_file_path = 'BF.zip'

# Extraction du contenu du fichier zip dans le même répertoire
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall()

# Vérifier l'existence du fichier extrait
extracted_files = os.listdir()
if 'BF.txt' in extracted_files:
    print("Extraction du fichier BF.txt réussie.")
else:
    print("Le fichier extrait BF.txt n'existe pas.")
```

```python
# Charger les données depuis le fichier extrait
data = pd.read_csv('BF.txt', sep='\t', header=None)
```

### 4 - Opérations sur le fichier CSV burkina_location.csv.

- extraire les données contenant le nom 'gounghin', enregistrez-le sous le fichier gounghin.csv
- extraire la sous-partie de la base de données (fichier burkina_location.csv), dont les noms les premières lettres des noms de lieux sont compris entre 'A' et 'P' (ordre alphabétique)
- Identifiez respectivement, la latitude, la longitude minimale et les noms de lieux correspondants
- Quels sont les lieux dont les coordonnées sont comprises entre (lat >= 11 et lon <= 0.5)


```python
#4.
# Charger les données depuis le fichier burkina_location.csv
burkina_data = pd.read_csv('burkina_location.csv')
```

```python
#extraire les données contenant le nom 'gounghin', enregistrez-le sous le fichier gounghin.csv
gounghin_data = burkina_data[burkina_data['location_name'].str.lower() == 'gounghin']
```

```python
# Enregistrer les données extraites dans un fichier CSV
gounghin_data.to_csv('gounghin.csv', index=False)
```

```python
# Extraire les données avec des noms de lieux dont les premières lettres sont entre 'A' et 'P'
sous_partie = burkina_data[(burkina_data['location_name'].str[0].str.upper() >= 'A') &
    (burkina_data['location_name'].str[0].str.upper() <= 'P')
]
```

```python
sous_partie
```

```python
# Trions les données par ordre alphabétique des noms de lieux
sous_partie = sous_partie.sort_values(by='location_name')
```

```python
sous_partie
```

```python
# Enregistrer les données extraites dans un fichier CSV
sous_partie.to_csv('sous_partie_location.csv', index=False)
```

```python

#Identifiez respectivement, la latitude, la longitude minimale et les noms de lieux correspondants
```

```python
#lecture
#sous_partie_data = pd.read_csv('sous_partie_location.csv')

# Trouvons la latitude minimale
min_latitude = burkina_data['lat'].min()

# Trouver la longitude minimale
min_longitude = burkina_data['long'].min()

# Trouver les noms de lieux correspondants à la latitude et longitude minimales
lieux_min_coordonnees = burkina_data[
    (burkina_data['lat'] == min_latitude) &
    (burkina_data['long'] == min_longitude)
]

# Afficher les résultats
print(f"Latitude minimale : {min_latitude}")
print(f"Longitude minimale : {min_longitude}")
print("\nNoms de lieux correspondants à la latitude et longitude minimales:")
print(lieux_min_coordonnees[['location_name', 'lat', 'long']])
```

```python

# Quels sont les lieux dont les coordonnées sont comprises entre (lat >= 11 et lon <= 0.5)
data_filtree = burkina_data[(burkina_data['lat'] >= 11) & (burkina_data['long'] <= 0.5)]
```
